```python
import os
import sys
import pandas as pd
import json
from pathlib import Path

sys.path.append(os.path.abspath(os.path.join('..')))

from api_clients import UniProtAPIClient
```

```python
# Directories and filenames
# Project directory --> Parent directory of the notebook
project_dir = Path(os.getcwd()).parent.parent
external_data_dir = os.path.join(project_dir, "external_data")
output_dir = os.path.join(project_dir, "output")
os.makedirs(output_dir, exist_ok=True)

gut_proteins_file = os.path.join(external_data_dir, "gut_proteins.json")
```

```python
# Load Gut Proteins
with open(gut_proteins_file, 'r') as f:
    gut_proteins = json.load(f)
gut_proteins = gut_proteins.keys()
gut_proteins = list(gut_proteins)
print("Number of Gut Proteins:", len(gut_proteins))

up_client = UniProtAPIClient()

response = up_client.batch_search(from_db="UniProtKB_AC-ID", to_db="UniProtKB", ids=gut_proteins, outdir=os.path.join(output_dir, "gut_proteins"))

print("Results:", response)
    
```
